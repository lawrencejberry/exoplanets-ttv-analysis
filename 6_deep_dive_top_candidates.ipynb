## Let's do a deep dive into the top three candidates

### Literature survey

#### Ivshina and Winn 2022:

Fitted constant period models using least squares, and then used the chi-squared test (i.e. sum of residuals) to determine ones might exhibit transit timing variation. Any system exceeding a chi-squared value of Ndof + 3*sqrt(2Ndof) was flagged as a candidate for TTV.

Then fitted decay model and flagged any dP/dE greater than 3 sd away from 0.

To avoid overfitting to outliers:
* visual inspection to identify any patterns in the residuals of the constant period least squares fit, such as oscillations, as well as isolated outliers.
* repeated the constant period least squares fit but assuming a minimum uncertainty in each transit time of 26 seconds and removing any >5sigma outliers.
* variable sigma0 error that gets added to measurements to represent different levels of systematic error

" First, to avoid missing anything obviously interesting, we visually inspected all of the timing residuals regardless of χ2, to look for any patterns in the residuals such as a grad- ual trend or an oscillation, as opposed to isolated outliers."

#### Hagey 2022: (TTV survey of orbital decay)

Bayesian posterior MCMC fitting of decay model, with BIC comparison of individual models. (Note to self: BIC is a heuristic, not a theoretical solution. Also, multiple physical phenomena could occur at the same time e.g. apsidal precession AND orbital decay).

To avoid overfitting:
* iterative sigma-clipping algorithm: fit the full model, then discard any > 3 sd outliers and refit. Repeat until no points fall in the 3 sd range.

"A worry of using the sigma-clipping method is that it may
introduce a false signal, a potential issue that was recognized early on. To address this, during development of the fitting routines, the same overall method was tried but with a best-fit linear model subtracted from the data instead of the best-fit orbital decay model."

"As a cautionary approach, the secondary analysis was repeated under the assumption that the transit center uncertain- ties in the ETD are underestimated, thus replacing them with the standard deviation of the spread of the timing residuals— i.e., the observed transit centers minus the times calculated from the best-fit linear ephemeris. In general, using a nonlinear least-squares algorithm can lead to optimistic parameter uncertainties."

#### Kokori et al. 2021: (updating ephemerides for Ariel scheduling)

Bayesian posterior MCMC fitting of a CONSTANT model. This is for computing accurate ephemerides of Ariel targets, not really for analysing decay.

"After a first fit, we scaled-up the uncertainties so that the mean uncertainty was equal to the STD of the O-C residuals and in this way incorporate any non-gaussian noise. While, this step did not have a significant effect on the values of the zero-epoch mid- transit and the period, it was important for their un- certainties. Without scaling, the uncertainties on the fi- nal ephemerides would had been largely underestimated, leading to reduced chi-squared values larger than 1."

"From the autocorrelation test we find that three plan- ets – WASP-12b, WASP-10b, and Qatar-2b – have a statistically significant O-C autocorrelation (> 3σ devi- ation from Gaussian noise) that indicates a non-linear ephemeris"

"The Shapiro test can help us identify high frequency noise in O-C diagrams"

#### Shan et al. 2022: (comparing TESS transit times to archival ephemeris)

Shan et al. employ the approach of comparing newly fitted TESS light curve transit times to that predicted by the current NASA exoplanet archive constant period parameters.

"TTV could originate from tidal dissipation, orbital precession, Rømer effect, mass loss and multi-planets"

"The timing uncertainty is derived as the quadratic sum of uncertainties of previous ephemeris and TESS timing."

"Type I target refers to a source of which timings are modeled with a lin- ear function. The timing inconsistency could be either due to systematic error underestimation or some physical process. The linear function indicates a model with a constant derivative, referring to a constant period. Type II refers to the tar- gets of which the timing differences can not be modeled by a linear function but by a quadratic function. The quadratic function can be due to abnormal points or physical processes which lead to a constant period derivative. We identify the targets as type III if the timings can not be fitted with any linear or quadratic functions. The possible physical origin of the timing offsets is discussed in Section.4."

"WASP-161b, XO-3b, and KELT-18b present timing off- sets larger than 10 σ. These three targets all have an ear- lier observed timing than the predictions from the previous ephemeris under a constant period assumption. We find WASP-161b and XO-3b present evidence of period decay." i.e. more than 10 sd away from the constant period (with sd of period from the NASA exoplanet archive).

### Checking for outliers

Plot TTV graphs for my top three candidates and plot the 3 sd bands (i.e. 3sd lower bound T0 + 3sd lower bound P0*E + 0.5* 3sd lower (most negative) bound dP/dE * E^2 and in this way check for any 3 sd outliers in the data:

```python
def fit_const_lst_sq_model(observations):
    a = np.array([np.ones(len(observations)), observations.index.values]).transpose()
    b = observations.values
    x, _, _, _ = sp.linalg.lstsq(a, b)
    lst_T0, lst_P0 = x[0], x[1]
    return lst_T0, lst_P0

def fit_analytical_posterior(observations, sd):
    # compute the analytical posterior of the observations
    sd2 = np.power(sd, -2)
    E = observations.index.values.astype(np.float64)
    T = observations.values.astype(np.float64)
    N = sd2.sum()

    a = (sd2*E).sum()
    b = (sd2*(E**2)).sum()
    c = (sd2*((E**3))).sum()
    d = (sd2*((E**4))).sum()

    A = (sd2*T).sum()
    B = (sd2*T*E).sum()
    C = (sd2*T*(E**2)).sum()

    N1 = a*c-b**2
    N2 = -(N*c-a*b)
    N3 = N*b-a**2
    DET = (d*(N*b-a**2)-b**3+2*a*b*c-N*c**2)
    
    mu_0 = (-d*b*A + (c**2)*A - a*c*C + (b**2)*C + d*a*B - b*B*c) / -DET
    var_0 = ((c**2)-d*b)/-DET
    
    mu_1 = (d*a*A - b*c*A - a*b*C + c*N*C + (b**2)*B - d*B*N) / -DET  # days
    var_1 =  ((b**2)-d*N)/-DET

    mu_2 = (N1*A+N2*B+N3*C)/DET
    mu_dP_dE = 2 * mu_2
    mu_dP_dE_ms = mu_dP_dE * (24 * 60 * 60 * 1000)
    
    var_2 = (N*b-(a**2))/DET  # good reason for this: must invert sigma^-1
    var_dP_dE = 4 * var_2
    var_dP_dE_ms = var_dP_dE * (24 * 60 * 60 * 1000)**2
    
    return mu_0, var_0, mu_1, var_1, mu_dP_dE_ms, var_dP_dE_ms
```

```python
def model(E, T0, P0, dP_dE):
    return T0 + P0*E + 0.5*(dP_dE/1000/60/60/24)*E**2

def fitting_method(file):
    df = pd.read_csv(file, sep="\t", names=["index", "epoch", "transit time", "error", "data quality", "observer"])
    df["transit time"] = df["transit time"] - df["transit time"].quantile(0.5)
    df["epoch"] = df["epoch"] - df["epoch"].quantile(0.5)

    observations = df["transit time"]
    observations.index = df.epoch.astype(int)
    error = df["error"].values

    lst_T0, lst_P = fit_const_lst_sq_model(observations)

    mu_T0, var_T0, mu_P, var_P, mu_dP_dE_ms, var_dP_dE_ms = fit_analytical_posterior(observations, sd=error)
    name = re.search("ETD_(.*)_og.txt", file).group(1)
    prob_dP_dE_less_than_0 = sp.stats.norm.cdf(x=0, loc=mu_dP_dE_ms, scale=np.sqrt(var_dP_dE_ms))
    
    E = observations.index
    const_y = model(E, lst_T0, lst_P, 0)
    decay_y = model(E, mu_T0, mu_P, mu_dP_dE_ms)
    three_sd = model(E, 3*np.sqrt(var_T0), 3*np.sqrt(var_P), 3*np.sqrt(var_dP_dE_ms))

    # outlier removal
    outliers = np.abs(observations.values - decay_y) > three_sd
    observations_wo_outliers = observations[~outliers]
    error_wo_outliers = df.error[~outliers].values
    
    lst_T0, lst_P = fit_const_lst_sq_model(observations_wo_outliers)
    mu_T0, var_T0, mu_P, var_P, mu_dP_dE_ms, var_dP_dE_ms = fit_analytical_posterior(observations_wo_outliers, sd=error_wo_outliers)
    name = re.search("ETD_(.*)_og.txt", file).group(1)
    prob_dP_dE_less_than_0 = sp.stats.norm.cdf(x=0, loc=mu_dP_dE_ms, scale=np.sqrt(var_dP_dE_ms))
    
    return name, mu_T0, var_T0, mu_P, var_P, mu_dP_dE_ms, var_dP_dE_ms, prob_dP_dE_less_than_0
```

```python
# running for all planets
files = glob.glob("./data/PROCESSED/*.txt")

results = []

for f in tqdm(files):
    name, mu_T0, var_T0, mu_P, var_P, mu_dP_dE_ms, var_dP_dE_ms, prob_dP_dE_less_than_0 = fitting_method(f)
    results.append([name, mu_T0, var_T0, mu_P, var_P, mu_dP_dE_ms, var_dP_dE_ms, prob_dP_dE_less_than_0])
    
results = pd.DataFrame(data=results, columns=["planet", "mu_T0", "var_T0", "mu_P", "var_P", "mu_dP_dE_ms", "var_dP_dE_ms", "Prob dP/dE < 0"])

results.sort_values(by="Prob dP/dE < 0", ascending=False)
results.to_csv("./results/6_analytical_fit_all_planets.csv")
```

### Improving outlier detection

Filtering outliers using standard deviations away from a fitted orbital decay model is problematic, as it has the potential to induce a false signal in the data by removing any data-points that appear to strongly disagree with the quadratic decay fit.

Filtering according to standard deviations from a constant fit may be more appropriate. Hagey tried this approach to double check the quadratic period fit wasn't inducing a false signal and found that fits where the linear model was better (according to BIC) resulted in quadratic fits with a negligible decay term i.e. clipping according to the quadratic model did not induce a false decay signal in models whose linear fit was actually better according to BIC.

However, we could equally lose indicators of a quadratic trend if clipping outliers according to deviation from a linear fit.

I want to build an outlier detection method centred around two ideas:

1. Focusing on physical intuition as means of detecting outliers i.e. can we use theory to predict possible "reasonable" bounds for what the period and decay rate could be, and use these to inform our modelling. We shouldn't try to fit to something which is clearly physically impossible. (See the HAT-P-1 example below.) => **Priors: introduce priors on the decay rate and period that force the model into a physical range.**

2. We can't be sure that the $1\sigma$ uncertainty of transit times given by the observers is correct. There are myriad reasons this uncertainty could be inaccurate, from systematic bias to human error. We need a way to accomodate "flexible" uncertainties in our model. Rather than simply clipping out bad data, which is quite a blunt instrument, we should be able to "de-weight" data-points dynamically. => **Dynamic uncertainty: make the uncertainty per transit a parameter of the model which we fit. The uncertainty provided in the data becomes our prior on this parameterized uncertainty instead.**
